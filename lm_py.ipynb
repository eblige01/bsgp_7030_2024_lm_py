## Reading in data

```python
import pandas as pd
data = pd.read_csv('regrex1.csv')
```

## Creating the scatterplot

```python
import matplotlib.pyplot as plt
x = data['x']
y = data['y']
plt.scatter(x,y)
plt.xlabel('X')
plt.ylabel('Y')           
plt.show()
```

## Model the data (linear model)

```python
import numpy as np
arrayX = np.array(x).reshape((-1,1))
arrayY = np.array(y)
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(arrayX,arrayY)
y_pred = model.predict(arrayX)
r_sq = model.score(arrayX,arrayY)
```

# Ploting model with the original data

```python
plt.scatter(x,y)
plt.plot(x,y_pred)
plt.xlabel('X')
plt.ylabel('Y')  
plt.show()
```
